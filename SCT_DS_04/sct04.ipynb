```python
import pandas as pd

data = {
    "Date": ["2024-01-05","2024-01-05","2024-01-06","2024-01-06","2024-01-07","2024-01-07","2024-01-08","2024-01-08"],
    "Time": ["08:30","18:20","14:10","22:40","09:15","17:50","11:30","20:10"],
    "Weather_Condition": ["Clear","Rain","Fog","Clear","Rain","Snow","Clear","Rain"],
    "Road_Condition": ["Dry","Wet","Wet","Dry","Icy","Snow","Dry","Wet"],
    "Accident_Severity": ["Low","High","Medium","Low","High","High","Low","Medium"],
    "Latitude": [12.9716,12.9720,12.9735,12.9700,12.9690,12.9750,12.9680,12.9710],
    "Longitude": [77.5946,77.5960,77.5980,77.5920,77.5900,77.5990,77.5890,77.5970]
}

df = pd.DataFrame(data)

df.to_csv("accident_data.csv", index=False)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Weather_Condition")
plt.title("Accidents by Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Road_Condition")
plt.title("Accidents by Road Condition")
plt.xlabel("Road Condition")
plt.ylabel("Count")
plt.show()
```

```python
df['Hour'] = pd.to_datetime(df['Time']).dt.hour

plt.figure(figsize=(10,5))
sns.histplot(df['Hour'], bins=24)
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour of Day (023)")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['Longitude'], df['Latitude'], s=50)
plt.title("Accident Hotspots")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
